```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
accuracies = []
f1_measure_list = []
precision_list = []  # List to store precision values
recall_list = []     # List to store recall values
num_iterations = 3

# Start the loop for a certain number of iterations
for i in range(num_iterations):
    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
    X_train = X_train.reset_index(drop=True)
    X_test = X_test.reset_index(drop=True)
    y_train = y_train.reset_index(drop=True)
    y_test = y_test.reset_index(drop=True)

    # Train the model using the current labeled data
    model = RandomForestClassifier()
    model.fit(X_train, y_train)
    uncertainty = model.predict_proba(X_test).max(axis=1)
        # Get the index of the most uncertain sample
    uncertain_sample = uncertainty.argmax()

    # Add the uncertain sample to the labeled data
    X_train = np.concatenate((X_train, [X_test.iloc[uncertain_sample]]))
    y_train = np.concatenate((y_test, [y_test.iloc[uncertain_sample]]))

    # Remove the uncertain sample from the test set
    X_test = X_test.drop(index=uncertain_sample)
    y_test.pop(uncertain_sample)

    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Calculate the accuracy of the model
    accuracy = accuracy_score(y_test, y_pred)
    f1_measure = f1_score(y_test, y_pred, average='weighted')
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')

    accuracies.append(accuracy)
    f1_measure_list.append(f1_measure)
    precision_list.append(precision)
    recall_list.append(recall)

    print(f"Iteration {i+1}: Accuracy = {accuracy:.4f}, F1 Score = {f1_measure:.4f}, Precision = {precision:.4f}, Recall = {recall:.4f}")

best_iteration = np.argmax(accuracies)
best_accuracy = accuracies[best_iteration]
print(f"Best Accuracy: {best_accuracy:.4f}")
```
